```python
#question 6:

class Queue:
    def __init__(self):
        self.items = []
        self.size = 0
 
    def is_empty(self):
        return self.items == []
 
    def enqueue(self, data):
        self.size += 1
        self.items.append(data)
 
    def dequeue(self):
        self.size -= 1
        return self.items.pop(0)
 
    def get_size(self):
        return self.size
    
    def dis(self):
        if self.size>0:
            print(self.items)


class Stack:
    def __init__(self):
        self.Q1 = Queue()
    
    def is_empty(self):
        return self.Q1.is_empty()
    
    def push(self, data):
        self.Q1.enqueue(data)
    
    def pop(self):
        if not self.Q1.is_empty():
            for _ in range(self.Q1.get_size() - 1):
                dequeued = self.Q1.dequeue()
                self.Q1.enqueue(dequeued)
            return self.Q1.dequeue()
        else:
            return "empty queue"
    
    def displayy(self):
        if not (self.Q1.is_empty()):
            self.Q1.dis()
        else:
            print("empty stack")
    
s=Stack()
s.push(7)
s.push(8)
s.push(9)
s.displayy()
print("popped: ",s.pop())
print("poped: ",s.pop())
print("poped: ",s.pop())

s.displayy()
```
